```python
import matplotlib.pyplot as plt
import numpy as np
```

# Partes de uma figura



# Grfico de Linhas

```python
# Exemplo simples
fig, ax = plt.subplots()             # Create a figure containing a single Axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the Axes.
ax.minorticks_on()
ax.set_xlabel("Valor")
ax.set_ylabel("Quantidade")
# ax.xaxis
plt.show()                           # Show the figure.
```

# Scatter

```python
x = np.random.rand(50)  # 50 pontos aleatrios para o eixo X
y = np.random.rand(50)  # 50 pontos aleatrios para o eixo Y
plt.scatter(x, y, color='blue', marker='o', label='Pontos')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Exemplo de Scatter Plot')
plt.legend()
plt.grid(False)

# Mostrar o grfico
plt.show()
```

## Dicas:

 Quando usar cada um?
- Se for um grfico simples e rpido  plt.plot()

 Vantagens:
Cdigo mais curto e rpido para grficos simples.
Fcil de usar quando s h um nico grfico na figura.

 Desvantagens:
Difcil de modificar aspectos avanados do grfico.
Menos flexvel para mltiplos grficos na mesma figura.


- Se precisar de personalizao avanada ou mltiplos grficos  fig, ax = plt.subplots()

 Vantagens:
Melhor controle sobre a figura e os eixos.
Essencial para criar mltiplos grficos na mesma figura (subplots).
Permite personalizaes mais avanadas.

 Desvantagens:
Cdigo um pouco mais longo e complexo para grficos simples.

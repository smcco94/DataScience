```python
import matplotlib.pyplot as plt
import numpy as np
```

# Partes de uma figura



# Gráfico de Linhas

```python
# Exemplo simples
fig, ax = plt.subplots()             # Create a figure containing a single Axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the Axes.
ax.minorticks_on()
ax.set_xlabel("Valor")
ax.set_ylabel("Quantidade")
# ax.xaxis
plt.show()                           # Show the figure.
```

# Scatter

```python
x = np.random.rand(50)  # 50 pontos aleatórios para o eixo X
y = np.random.rand(50)  # 50 pontos aleatórios para o eixo Y
plt.scatter(x, y, color='blue', marker='o', label='Pontos')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Exemplo de Scatter Plot')
plt.legend()
plt.grid(False)

# Mostrar o gráfico
plt.show()
```

## Dicas:

🔥 Quando usar cada um?
- Se for um gráfico simples e rápido → plt.plot()

✅ Vantagens:
Código mais curto e rápido para gráficos simples.
Fácil de usar quando só há um único gráfico na figura.

🚫 Desvantagens:
Difícil de modificar aspectos avançados do gráfico.
Menos flexível para múltiplos gráficos na mesma figura.


- Se precisar de personalização avançada ou múltiplos gráficos → fig, ax = plt.subplots()

✅ Vantagens:
Melhor controle sobre a figura e os eixos.
Essencial para criar múltiplos gráficos na mesma figura (subplots).
Permite personalizações mais avançadas.

🚫 Desvantagens:
Código um pouco mais longo e complexo para gráficos simples.

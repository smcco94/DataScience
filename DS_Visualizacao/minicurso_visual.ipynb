```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gráficos de Linha
x = [1,2,5]
y = [2,4,6]

plt.title("Grafico de Linha")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.minorticks_on()
plt.plot(x,y)
```

```python
# Gráficos de Barras
x = [1,2,3,4,5]
y = [2,4,6,1,17]

plt.title("Grafico de Barras")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.bar(x,y)
```

```python
# Gráficos de Barras Comparado
x1 = [1,3,5,7,9]
y1 = [2,4,6,1,17]

x2 = [0,2,4,6,8]
y2 = [4,4,12,4,5]

plt.title("Grafico de Barras")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.bar(x1,y1, label="Grupo 1")
plt.bar(x2,y2, label="Grupo 2")
plt.legend()
```

```python
# Grafico de Dispersão

# Gerando dados aleatórios
x = np.random.rand(20)  # 50 valores aleatórios para X
y = np.random.rand(20)  # 50 valores aleatórios para Y

# Criando o scatter plot
plt.scatter(x, y, color="blue", marker="o", alpha=0.6, label="Pontos")

# Personalizando o gráfico
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Exemplo de Scatter Plot")
plt.legend()
plt.grid(True)

# Caso queira salvar:
# plt.savefig("teste.png", dpi=300)
# plt.savefig("teste.pdf")
```

```python
# Boxplot

# Gerando dados aleatórios
np.random.seed(42)
dados = np.random.normal(loc=50, scale=15, size=100)  # Média 50, desvio padrão 15

# Calculando o IQR (Intervalo Interquartil)
Q1 = np.percentile(dados, 25)  # Primeiro quartil (25%)
Q3 = np.percentile(dados, 75)  # Terceiro quartil (75%)
IQR = Q3 - Q1  # Intervalo Interquartil
limite_inferior = Q1 - 1.2 * IQR
limite_superior = Q3 + 1.2 * IQR

# Removendo os outliers
dados_filtrados = dados[(dados >= limite_inferior) & (dados <= limite_superior)]

# Criando o Boxplot sem outliers
plt.boxplot(dados_filtrados, vert=True, patch_artist=True, boxprops=dict(facecolor="lightblue"))

# Personalizando o gráfico
plt.title("Boxplot sem Outliers (IQR * 1.2)")
plt.ylabel("Valores")
plt.grid(True, linestyle="--", alpha=0.6)

# Exibir o gráfico
plt.show()
```
